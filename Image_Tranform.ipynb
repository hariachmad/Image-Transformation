```python
import cv2 

pathImg="./Image/Ronaldo3.png"
img=cv2.imread(pathImg)


```

```python
cv2.imshow('image',img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
#Melakukan Transformasi Scale Gambar
img_small_area=cv2.resize(img,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_AREA)
img_small_cubic=cv2.resize(img,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_CUBIC)
img_big_cubic=cv2.resize(img,None,fx=1.5,fy=1.5,interpolation=cv2.INTER_CUBIC)
img_big_area=cv2.resize(img,None,fx=1.5,fy=1.5,interpolation=cv2.INTER_AREA)
while(1):
    cv2.imshow('image asli',img)
    cv2.imshow('image small area',img_small_area)
    cv2.imshow('image small cubic',img_small_cubic)
    cv2.imshow('image big cubic',img_big_cubic)
    cv2.imshow('image big area',img_big_area)
    if cv2.waitKey(100) & 0xFF == 27:
        break           
cv2.destroyAllWindows()

#Memperkecil gambar menggunakan interpolation Area
#Memperbesar gambar menggunakan interpolation Cubic
```

```python
print(img.shape)
print(img_big_area.shape)
print(img_small_area.shape)
```

```python
rows,cols,dims=img.shape
m90=cv2.getRotationMatrix2D((rows/2,cols/2),90,1)
img_90=cv2.warpAffine(img,m90,(cols,rows))
cv2.imshow('90 Derajat',img_90)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
